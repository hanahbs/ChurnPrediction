```python
import pandas as pd
df = pd.read_csv("churnprediction_ch9.csv",sep=',',index_col=['customer_id'])
```

```python
df.describe()
```

```python
df_aktif_churn = df.groupby('churn').count()
```

```python
import matplotlib.pyplot as plt
plt.pie(df_aktif_churn['product'],labels=['Aktif','Churn'],autopct='%1.0f%%')
plt.title("Persentase Pelanggan Aktif vs Churn")
plt.axis("equal")
plt.show()
```

```python
df2 = pd.concat([df,pd.get_dummies(df['product'])],axis=1,sort=False)
df2.drop(['product'],axis=1,inplace=True)
```

```python
df2
```

```python
# memvisualisasikan korelasi dengan nilai korelasi yg tinggi ditandai dengan warna yang lebih terang
import seaborn as sns

dfKorelasi = df2.corr()
sns.heatmap(dfKorelasi, 
            xticklabels=dfKorelasi.columns.values,
            yticklabels=dfKorelasi.columns.values,
            annot=True, 
            annot_kws={'size': 12})
# gcf = ketika Anda perlu melakukan modifikasi lebih lanjut pada figure yang sudah ada setelah plot dibuat.
heat_map = plt.gcf()
heat_map.set_size_inches(10,10)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.show()
```

```python
X= df2.drop(['reload_2','socmed_2','games','churn'],axis=1,inplace=False)
y= df2['churn']
```

```python
import sklearn.model_selection as ms
X_train,X_test,y_train,y_test = ms.train_test_split(X,y,test_size=0.8,random_state=0)
```

```python
import sklearn.preprocessing as pp
scl = pp.StandardScaler(copy=True,with_mean=True,with_std=True)
scl.fit(X_train)
X_train=scl.transform(X_train)
X_test=scl.transform(X_test)
```

```python
import sklearn.model_selection as ms
import sklearn.linear_model as lm
import sklearn.metrics as met
# lbfgs adalah salah satu metode
model = lm.LogisticRegression(solver='lbfgs')
model.fit(X_train,y_train)
```

```python
y_prediksi = model.predict(X_test)
print(y_prediksi)
score = met.accuracy_score(y_test,y_prediksi)
print("accuracy=",score)
precision = met.precision_score(y_test,y_prediksi)
print("precision=",precision)
recall=met.recall_score(y_test,y_prediksi)
print("recall=",recall)
auc=met.roc_auc_score(y_test,y_prediksi)
print("auc=",auc)
```

```python
import sklearn.ensemble as ens
# Menentukan jumlah pohon keputusan dalam hutan
model = ens.RandomForestClassifier(n_estimators=200,random_state=0)
model.fit(X_train,y_train)
```

```python
y_prediksi = model.predict(X_test)
print(y_prediksi)
score=met.accuracy_score(y_test,y_prediksi)
print("accuracy=",score)
precision=met.precision_score(y_test,y_prediksi)
print("precision=",precision)
recall=met.recall_score(y_test,y_prediksi)
print("recall=",recall)
auc=met.roc_auc_score(y_test,y_prediksi)
print("auc=",auc)
```

```python
feature_penting = pd.Series(model.feature_importances_,index=X.columns)
# barh = grafik batang horizontal
feature_penting.nlargest(10).plot(kind='barh')
```
